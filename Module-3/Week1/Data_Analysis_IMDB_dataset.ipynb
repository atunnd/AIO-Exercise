Step 5: Data Selection – Based on Conditional filtering

```python
df[((df['Year'] >= 2010) & (df['Year'] <= 2015))
   & (df['Rating'] < 6.0)
   & (df['Revenue (Millions)'] > df['Revenue (Millions)'].quantile(0.5))].head(5)
```

Step 6: Groupby operations

```python
# Finding the mean rating of each director
df.groupby('Director')['Rating'].mean().head(10)
```

Step 7: Sorting operation

```python
# Sorting the directors from high ratings to low ratings
df.groupby('Director')['Rating'].mean().sort_values(ascending=False)
```

Step 8: View missing values

```python
# view missing values by count number of missing values
df.isnull().sum()
```

Step 9: Deal with missing values - Deleting

```python
# Drop columns containing many missing values
df.drop(columns='Metascore', inplace=True)
df.info()
```

Step 10: Deal with missing values - Filling

```python
# Filling values with mean value
df.fillna({'Revenue (Millions)': df['Revenue (Millions)'].mean()}, inplace=True)
df.info()
```

Step 11: Apply() functions

```python
# Apply functions to categorize film into 3 sections: 'Good', 'Average', and 'Bad'
def rating_group(rating):
    if rating < 6.0:
        return 'Bad'
    elif rating < 7.5:
        return 'Average'
    else:
        return 'Good'

df['Rating_category'] = df['Rating'].apply(rating_group)
df[['Title', 'Genre', 'Director', 'Year', 'Rating_category']].head(5)
```
