```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load the diabetes dataset
iris_X, iris_y = datasets.load_iris(return_X_y=True)
```

```python
# Spit train:test = 8:2
X_train, X_test, y_train, y_test = train_test_split(
    iris_X,
    iris_y,
    test_size=0.2,
    random_state=42
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train, y_train)
```

```python
y_pred = knn_classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
plt.figure(figsize=(10,5)) # Create a figure object

plt.subplot(1, 2, 1)
plt.scatter(range(len(y_test)), y_test, color='blue', label='y_test')
plt.scatter(range(len(y_pred)), y_pred, color='red', marker='x', label='y_pred')
plt.title('Comparison of y_test and y_pred')
plt.xlabel('Index')
plt.ylabel('Class')
plt.legend()
```
